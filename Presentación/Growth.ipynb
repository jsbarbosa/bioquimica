```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("growth.csv")
df
```

```python
x = "Year"

y_text = ["XRAY", "NMR", "EM"]
labels = ["Rayos X", "RMN", "ME-Crio"]

fig, ax = plt.subplots(figsize = (8, 4.5))

for (y, l) in zip(y_text, labels):
    ax.bar(df[x], df["Released Annually " + y], alpha = 0.5, edgecolor = "k", label = l)
    
ax.set_yscale('log')

ax.set_xlabel("Año")
ax.set_ylabel("Estructuras publicadas por año")

ax.legend()
fig.tight_layout()

fig.savefig("growth.png", dpi = 300)
```
